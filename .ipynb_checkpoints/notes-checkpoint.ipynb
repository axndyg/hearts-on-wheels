### Fill dummy data

numpy is a core python package offering python code wrapped over pre-compiled C; this provides

> Concise Code <br> 
> Efficient execution

The core being a **ndarray** object 

```
for i in range(len(c): 
    c.append(a[i] * b[i])
```

If `a`, `b`, and `c`, were instances of a **ndarray**, this simplifies into 

```
c = a * b
```

With the same functionality being implied if `a` and `b` were instances of like-sized 2D-arrays

````python
import numpy as np

dummy_a = np.arange(1,4)
print(dummy_a)
````

````python
dummy_ba = np.arange(4,8)
dummy_bb = np.arange(8,12)
dummy_b = np.vstack((dummy_ba.flatten(), dummy_bb.flatten())).T
ind_b = ['a', 'b', 'c', 'd']
print(dummy_b)
````

````python
rws, cols = 3, 5
ind_c = ['a', 'b', 'c']
dummy_c = np.random.rand(rws, cols)
dummy_c = dummy_c * 10
dummy_c = dummy_c.round(2)
print(dummy_c)
````

````python
# Creation Method One 
df_a = pd.DataFrame(
    data=dummy_a,
)

df_b = pd.DataFrame(
    data=dummy_b,
)

## Creation Method Two 
df_c = pd.DataFrame.from_records(dummy_c)
df_c.index = ind_c
````

### Plot with pure matplotlib

Plotting with **matplot**, or "plt," is more verbose, or rather there is stricter placements on where settings are listed.
**df.matplot** plotting offers more flexibiltiy on setting placement

```
    plt.bar() vs. df.plot.bar() 
                  df.plot(kind="bar")    
```

JuptyerLab shows a report of the variable last executed, this explains the differente between 

```
     print(a) vs a 
```

When using **matplot**, the equivalent of `print(a)` would then be `plt.show()`

````python
import matplotlib.pyplot as plt 

x=df_c["a"]
y=df_c["b"]
plt.plot(x, y,
            #**kwargs
            linewidth=2,
            color="orange",
        )
plt.plot(y,x)
plt.show()

plt.bar(y, x,
            #**kwargs
            linewidth=0.5,
            color="red",
        )
plt.show()
````

`!` **Notice above** 

`plt` is the occurence of the **matplotlib** library, and we don't actually instantiate a plot object. Rather we configure the plot library with multiple settings and it creates a figure on a window

> This is why the first plot overlaps our first two `plt.plot()` calls, <br>
> we are drawing two figures to the same window

`plt.show` will display the drawing and consequently clear the figure settings we wrote 

> This is why the bar graph is generated in its own window. <br>
> Notice the order of `plt.plot()` and `plt.show()`



#### Subplots

But what if we want tow related figures on the same window, but seprate graphs? 

Then we have **subplots**. Subplots are entirely possilbe for both **df.matplot** and **matplot**, but it is more convienent for formatting reasons to establish subplots with **matplot**

````python
import matplotlib.pyplot as plt 

nrows = 2
ncols = 3 
fig, axes = plt.subplots(nrows, ncols)

    # fig is the window
    # axes is a graph on the window, indexed axes[

x = df_b[0]
y = df_b[1]
axes[0, 0].plot(x,y)
axes[0, 1].bar(y,x)

axes[0, 2].stackplot(x,y)
axes[0, 2].stackplot(x * .8, y / 2, color="red")

axes[1,0].fill_between(x,y)
axes[1,1].stem(x,y)
axes[1,2].hist(y, bins=6)

plt.subplots_adjust(
    left=.8,
    right=2,
    hspace=.3,
    wspace=.5
)

plt.show()
````

`!` **Learning Above**

The `plt.subplots()` functionality returns two objects 

1. `fig ` : the window itself generated, also visible with plt.show() 
1. `axes` : dimenisional array holding the subplots of the figure,
    1. nrows = 1 = ncols, axes is a singular axis, not an array of axes
    1. nrows = 1: indexed with singular position, i.e `axes[r_i]`
    1. nrows > 1: index with coordinates, i.e `axes[r_i, c_i`]

### Plot with both 

````python
import matplotlib.pyplot as plt 

f_width = 10
f_height = 2
nrows = 1
ncols = 1 

fig, axes = plt.subplots(nrows, ncols, figsize=(f_width, f_height), dpi=150)

df_dm.plot(
    kind="scatter",
    ax=axes,
    x="body_mass_g",
    y="bill_depth_mm",
    label="male"
)
df_df.plot(
    kind="scatter",
    ax=axes,
    x="body_mass_g",
    y="bill_depth_mm",
    color="red",
    label="female"
)

axes.set_title("Penguin body mass (g) vs bill depth (mm), by sex")
axes.set_xlabel("body mass (g)")
axes.set_ylabel("bill depth (mm)")

plt.show()
````

`!` **Learning Above**

If we desire mixed functionality of **df.matplot** and **matplot**, we can make use of the ```plt.subplots()``` functionality.

```
fig, axes = plt.subplots(2, 2)

df_a.plot(
    x="x", y="y",
    kind=bar, ax=axes[0,0]
    )

df_b.plot(
    x="x", y="y",
    kind=bar, ax=axes[0,1]
    )
```

With the axes, we can set a plot's placement within our figure. If we have ```ncols = nrows = 1```, there is simply one mane axes on the figure

In this manner, we pair a dataframe's plot to an axes and can manipulate said plot outside ```df.plot()``` settings and mix and match to our conveinence
